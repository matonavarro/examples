### Data Cleansing

```python
#Let's read the full csv data set and display the top records
import pandas as pd
data = pd.read_csv(f'{path}/Customer-Churn.csv')
print(data.head())
```

##### After reading the data as a Pandas DataFrame we view a subset and determine that we do niot see any immediate issues.  We can now view the column datatypes and check for non-missing values

```python
#Let's get more info on each of the columns
data.info()
```

##### The TotalCharges column (which is a number) is stored as an object.  To perform numerical analysis on this column we must convert this to a float64.

```python
#Lets convert TotalCharges to a numeric data type
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'])
```

##### The ValueError above tells us that there is at least one string value  '  '  in the TotalCharges column that cannot pe parsed to numeric.  We can explore the data set further to identify any other records that contain ' '  as a TotalCharge value.

```python
#lets find all records that have an empty string value in TotalCharges
data.loc[data['TotalCharges'] == ' ']
```

##### We see that eleven records have ' ' as a TotalCharge value.  Viewing our resluts from above we can assume that the TotalCharges column is the sum of the MonthlyCharges column and other charges not explicitly recorded in this data set. Therefore, we can assume that ' ' actually means 0.00 and can therefore perform a replace on these values.

```python
#Let's replace ' ' with 0 and validate
data['TotalCharges'] = data['TotalCharges'].replace(' ', 0)
data.loc[data['TotalCharges'] == ' ']
```

```python
#Let's try converting Total Charges to numeric again
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'])
data.info()
```
